```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from rdkit.Chem import Draw
from random import shuffle

from data.utils import *
from data.visualize import *
```

```python
def get_mols(bonds, atoms):
    mols = []
    for bond, atom in zip(bonds, atoms):
        mol, smiles = get_rdkit_molecule(bond, atom)
        mols.append(mol)
    return list(filter(lambda m: m!=None, mols))
```

```python
bonds, atoms = read_molecules()
mols = get_mols(bonds, atoms)
```
